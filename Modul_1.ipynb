> Langkah (Step) 1:  
  
Pada code awal, gunakan beberapa library : numpy, pandas, cv2, skimage, PIL, matplotlib

```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pyplot as plt
```

> Langkah (Step) 2:

Baca Image dari URLs

```python
# Membuat list untuk menyimpan url dari beberapa image
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg","https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg","https://as1.ftcdn.net/v2/jpg/02/31/17/32/1000_F_231173210_stzZNH7tblr3esfej44EpbKqAQfbkfsT.jpg"]

# baca dan tampilkan image
# loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
  image = io.imread(url)                          #read image
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5) #resize image to half size
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)  #convert color to RGB
  final_frame = cv.hconcat([image, image_2])      #concateenate image
  cv2_imshow(final_frame)                         #show image
  print("\n")
```

> Langkah (Step) 3:

Melihat ukuran file image

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image: tinggi x lebar = ", tinggi, "x", lebar)
cv2_imshow
```

> Langkah (Step) 4:

Mengakses pixel dengan memberi garis diagonal menyilang

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis horizontal ditengah image
for y in range (lebar):
  image_3[int((tinggi)/2), y] = [255, 255, 255]

final_frame = cv.hconcat([image_2, image_3])
cv2_imshow(final_frame)
```

2. Buat garis Vertikal dan garis menyilang diagonal pada image keluaran

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_4 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis vertical ditengah image
for x in range (tinggi):
  image_4[x, int((lebar)/2)] = [255, 255, 255]

final_frame = cv.hconcat([image_2, image_4])
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_5 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# garis diagonal kiri-atas -> kanan-bawah
cv.line(image_5, (0, 0), (lebar-1, tinggi-1), (255, 255, 255))

# garis diagonal kanan-atas -> kiri-bawah
cv.line(image_5, (lebar-1, 0), (0, tinggi-1), (255, 255, 255))

final_frame = cv.hconcat([image_2, image_5])
cv2_imshow(final_frame)
```

3. Buat garis horizontal berwarna putih dibagian tengah gambar dengan panjang tertentu

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_6 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# koordinat y di tengah
y_tengah = tinggi // 2

# panjang garis (misalnya 200 pixel)
panjang_garis = 200
x_awal = (lebar - panjang_garis) // 2
x_akhir = x_awal + panjang_garis

# gambar garis horizontal putih di tengah
cv.line(image_6, (x_awal, y_tengah), (x_akhir, y_tengah), (255, 255, 255))

# tampilkan hasil
final_frame = cv.hconcat([image_2, image_6])
cv2_imshow(final_frame)
```

4. Buat kotak menggunakan kumpulan pixel warna putih di sembarang tempat dalam gambar

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_7 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# tentukan posisi kotak (x1,y1) sebagai pojok kiri-atas, (x2,y2) sebagai pojok kanan-bawah
x1, y1 = 150, 100      # sembarang koordinat
x2, y2 = 250, 200    # ukuran kotak (lebar 100px, tinggi 100px)

# buat kotak dengan pixel putih
image_7[y1:y2, x1:x2] = [255, 255, 255]

# tampilkan
final_frame = cv.hconcat([image_2, image_7])
cv2_imshow(final_frame)
```
